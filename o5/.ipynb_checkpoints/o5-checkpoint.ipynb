```python
# Standard imports for pandas and sklearn sample datasets
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets

# Load the mushroom data
mushroom_data = pd.read_csv("https://datahub.io/machine-learning/mushroom/r/mushroom.csv")

# Preview first 5 rows of imported data
mushroom_data.head()

# Convert it to a pandas DataFrame
#df = pd.DataFrame(mushroom_data, columns=mushroom_data.feature_names)

# A summary of the data
#df.describe()
```

```python
#We will compare just the habitat and edibility
filtered_data = mushroom_data.filter(["habitat", "class"])
#finds edible mushrooms
edible = filtered_data[filtered_data['class'] == 'e']
edible['habitat'].value_counts()
```

```python
#now compare them to poisonous
poisonous = filtered_data[filtered_data['class'] == 'p']
poisonous['habitat'].value_counts()
```

```python
#this will get the dummy vaiable version of the column
pd.get_dummies(mushroom_data.habitat)
```

```python
habitats = mushroom_data.filter(["habitat"])
out = pd.DataFrame(habitats['habitat'].value_counts().reset_index().values, columns=["Habitat", "Mushrooms"])

habitat_headers = mushroom_data.habitat.unique()
habitat_headers_ids = {}

for i, header in enumerate(habitat_headers):
    habitat_headers_ids[header] = i

size = mushroom_data.shape[0]
print(size)

habitat_features = np.zeros((size, len(habitat_headers)))
    
for i, row in mushroom_data.iterrows():
     habitat_features[i, habitat_headers_ids[row['habitat']]] = 1

plt.spy(habitat_features, aspect='auto', markersize=0.1)
fig = plt.gcf()

ax = plt.gca()

labels = [item.get_text() for item in ax.get_xticklabels()]
labels = habitat_headers
labels = np.insert(labels, 0, '-1', axis=0)

ax.set_xticklabels(labels)

fig.set_size_inches(20,5)
plt.plot()
plt.show();


```
